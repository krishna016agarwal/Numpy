## Advabce operation with Business example

```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
sales_data=np.array([
 [    1 ,51277 ,97091, 74155, 20324],
 [    2 ,69852 ,56674 ,94836 ,27187],
 [    3 ,10220 ,88056 ,10557 ,96218],
 [    4 ,87865, 93730 ,53185 ,66973],
 [    5 ,31045, 82642 ,19242, 56456],
 ])
```

```python
print("===Zomato sales analysis===")
print(sales_data.shape)
print("\nSample data for 1st 3 restudent \n",sales_data[:3])
```

### total sales per year

```python
print(np.sum(sales_data,axis=0))

yearly_total=np.sum(sales_data[:,1:],axis=0) # remove id , axis=0 means adding verically
print("Toatl sales of all restaurants in a year ",yearly_total)

#Minimum sales per restudent
min_sales=np.min(sales_data[:,1:],axis=1)
print("Minimum sales ",min_sales)

#Maximum sales from all resturants
max_sales=np.max(sales_data[:,1:],axis=0)
print("Maximum sales from all resturants ",max_sales)

#Average sales per resturants
avg_sales=np.mean(sales_data[:,1:],axis=1)
print("Average sales per resturants ",avg_sales)

cumulative_sales=np.cumsum(sales_data[:,1:],axis=1)
print(cumulative_sales)  #increased in sales , add current year sales to previous year sales

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumulative_sales,axis=0))
plt.title("Average cumulative sales accross all resturants ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Operations

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([7,8,9,4,56])
print(vector1+vector2)

print("Multiplication of vectors ",vector1*vector2)
print("Dot product ",np.dot(vector1,vector2))

angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle bwt vectors = ",angle)
```
